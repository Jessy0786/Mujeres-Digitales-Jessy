# TALLER 1
La empresa VIVAFLY tiene como eslogan: "Satisfacción y plenitud en un solo viaje". Viaje todo incluído para dos personas a San Andrés(Premio 1)

1. Rosado= Premio 1

3. Azul=Premio 3
4. Gris=Premio 4
5. Roja=No hay premio
