```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
def f(x):
    return np.exp(-x ** 2)

x = np.linspace(-5,5, num=10)

plt.plot(x,f(x),'y', linestyle='--', marker='o', label="f(x) Function")#color:'y', estilo de linea'--', marcado'o'
plt.ylabel('f(x)')
plt.xlabel('x')
plt.legend()
```

```python
plt.style.use('dark_background')#fondo negro
plt.plot(x,f(x),'y', linestyle='--', marker='o', label="f(x) Function")#color:'y', estilo de linea'--', marcado'o'
plt.ylabel('f(x)')
plt.xlabel('x')
plt.legend()
```

```python
plt.style.use('ggplot')#cuadricula gris
plt.plot(x,f(x),'y', linestyle='--', marker='o', label="f(x) Function")#color:'y', estilo de linea'--', marcado'o'
plt.ylabel('f(x)')
plt.xlabel('x')
plt.legend()
```
